```python
region_groups = {
    "India": "South Asia",
    "Pakistan": "South Asia",
    "Bangladesh": "South Asia",
    "China": "East Asia",
    "Japan": "East Asia",
    "Korea, Rep.": "East Asia",
    "Philippines": "Southeast Asia",
    "Indonesia": "Southeast Asia",
    "Viet Nam": "Southeast Asia"
}
```

```python
secondarymerged['Region'] = secondarymerged["Country Name"].map(region_groups)
```

```python
secondaryavgs = secondarymerged.groupby(["Region","Year"])["Enrollment Rate"].mean().reset_index()
```

```python
secondaryavgs
```

```python
import matplotlib.pyplot as plt

# Plot each country separately
plt.figure(figsize=(12, 6))
for region in secondaryavgs["Region"].unique():
    region_data = secondaryavgs[secondaryavgs["Region"] == region]
    plt.plot(region_data["Year"], region_data["Enrollment Rate"], label=region, marker="o")


# Add labels and title
plt.title("Secondary School Enrollment Rate by Year", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Enrollment Rate (%)", fontsize=14)
plt.legend(title="Country")
plt.grid(True)
plt.show()
```
